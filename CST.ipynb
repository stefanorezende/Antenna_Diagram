```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
import numpy as np
from scipy import interpolate
```

```python
df = pd.read_csv('CST\\2AHHF06.txt', sep='\s+', skiprows=2, names= ['Theta [deg.]','Phi [deg.]','Abs(Grlz)[dBi]',  'Abs(Theta)[dBi]',  'Phase(Theta)[deg.]', 'Abs(Phi)[dBi]', 'Phase(Phi)[deg.]', 'Ax.Ratio[dB]'])
df
```

```python
df['POSr'] = np.radians(range(90,450, 1))
df
```

```python
df = df.sort_values(by= 'POSr', ascending=True)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='polar')
ax.plot(df['POSr'], df['Abs(Theta)[dBi]'], label='Dados de Exemplo', color='blue')
plt.yticks(range(round(max(df['Abs(Theta)[dBi]'])-20), round(max(df['Abs(Theta)[dBi]'])), 3))

ax.set(ylim=(max(df['Abs(Theta)[dBi]'])-20, max(df['Abs(Theta)[dBi]'])))
ax.set_title('GrÃ¡fico Polar Personalizado')
ax.legend()
plt.show()
```
